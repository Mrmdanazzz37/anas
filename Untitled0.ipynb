```python
import heapq
import matplotlib.pyplot as plt
import networkx as nx

def dijkstra(graph, start):
    distance = {node: float('inf') for node in graph}  # Initialize distances to infinity
    distance[start] = 0  # Distance to the start node is 0
    queue = [(0, start)]  # Min-heap to store nodes with the current shortest distance

    while queue:
        curr_dist, node = heapq.heappop(queue)  # Pop the node with the smallest distance

        # Explore each neighbor of the current node
        for neighbor, weight in graph[node].items():
            dist = curr_dist + weight  # Calculate new distance through this neighbor
            if dist < distance[neighbor]:  # If a shorter path is found
                distance[neighbor] = dist  # Update the distance
                heapq.heappush(queue, (dist, neighbor))  # Push the neighbor to the queue

    return distance  # Return the shortest distances from the start node

def draw_graph(graph, shortest_paths=None, start_node=None):
    G = nx.Graph()  # Create an empty graph

    # Add edges to the graph with weights
    for node in graph:
        for neighbor, weight in graph[node].items():
            G.add_edge(node, neighbor, weight=weight)

    # Create a spring layout for positioning the nodes
    pos = nx.spring_layout(G)
    edge_labels = nx.get_edge_attributes(G, 'weight')  # Get edge weights

    # Draw the graph
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1000)
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

    # Highlight the shortest path edges (if provided)
    if shortest_paths:
        path_edges = []
        for node in shortest_paths:
            # Trace the shortest path from the start node to every other node
            for neighbor, weight in graph[node].items():
                if shortest_paths[node] + graph[node].get(neighbor, float('inf')) == shortest_paths[neighbor]:
                    path_edges.append((node, neighbor))  # Add to path_edges
        nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=2)

    # Title and display the plot
    plt.title(f"Shortest Path from {start_node}" if start_node else "Energy Optimization Graph")
    plt.show()

if __name__ == "__main__":
    # Define a weighted undirected graph
    graph = {
        'A': {'B': 2, 'C': 1, 'D': 1},
        'B': {'A': 2, 'C': 2, 'D': 1},
        'C': {'A': 1, 'B': 2, 'D': 2},
        'D': {'A': 1, 'B': 1, 'C': 2}
    }

    # Calculate shortest paths from node 'A'
    start_node = 'A'
    shortest_distances = dijkstra(graph, start_node)

    # Draw the graph with shortest paths highlighted
    draw_graph(graph, shortest_paths=shortest_distances, start_node=start_node)

    # Print shortest distances
    print(f"Shortest distances from node {start_node}:")
    for node, dist in shortest_distances.items():
        print(f"{node}: {dist}")
```
